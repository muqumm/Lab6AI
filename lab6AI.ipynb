```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
# Load the data from CSV
df = pd.read_csv('Book1.csv')
# Display the first few rows of the DataFrame (optional)
print(df.head())
# Plotting the data
plt.figure(figsize=(10, 6))
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df['Area'], df['Price'], color='red', marker='+', label='Data Points')
# Prepare the data for regression
new_df = df.drop('Price', axis='columns')
price = df['Price']
# Create and fit the regression model
reg = LinearRegression()
reg.fit(new_df, price)
# Predicting the price for a range of areas for plotting the regression line
area_range = np.linspace(df['Area'].min(), df['Area'].max(), 100).reshape(-1, 1)
predicted_prices = reg.predict(area_range)
# Plot the regression line
plt.plot(area_range, predicted_prices, color='blue', label='Regression Line')
# Predicting the price for a specific area
specific_area = 3300
predicted_price = reg.predict([[specific_area]])
print(f'Predicted price for area {specific_area}: {predicted_price[0]}')
# Displaying coefficients and intercept
print(f'Coefficient: {reg.coef_[0]}')
print(f'Intercept: {reg.intercept_}')
# Adding a legend and showing the plot
plt.legend()
plt.title('Area vs Price')
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
df = pd.read_csv('Book3.csv')
print(df.head())
X = df[['Horsepower', 'Mileage']]  # Features
y = df['Price']  # Target variable
reg = LinearRegression()
reg.fit(X, y)
specific_horsepower = 250
specific_mileage = 12000
predicted_price = reg.predict([[specific_horsepower, specific_mileage]])
print(f'Predicted price for car with {specific_horsepower} HP and {specific_mileage} miles: {predicted_price[0]}')
print(f'Coefficients: {reg.coef_}')
print(f'Intercept: {reg.intercept_}')
plt.figure(figsize=(10, 6))
plt.scatter(df['Horsepower'], y, color='blue', label='Actual Prices')  # Actual data points
plt.scatter(specific_horsepower, predicted_price, color='green', label='Predicted Price', marker='x', s=100)  # Specific prediction
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.title('Horsepower vs Price')
plt.legend()
plt.grid(True)
horsepower_range = np.linspace(df['Horsepower'].min(), df['Horsepower'].max(), 100).reshape(-1, 1)
predicted_prices = reg.predict(np.column_stack((horsepower_range, np.mean(df['Mileage']) * np.ones(horsepower_range.shape[0]))))  # Using average mileage for line
plt.plot(horsepower_range, predicted_prices, color='red', label='Regression Line')  # Regression line
plt.legend()
plt.show()
```
